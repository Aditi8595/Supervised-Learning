```python
#https://forms.gle/B6KxfZXjbKwJXXnEA


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.metrics as metrics

# Importing the datasets

datasets = pd.read_csv(r'C:\Users\HP\Downloads\bank-additional-full.csv')
datasets.head()
```

```python
datasets.info()
```

```python
category={'yes':1,'no':0,'unknown':2,'nonexistent':2,'failure':0,'success':1,'cellular':1,'telephone':0}
datasets['y']=datasets['y'].map(category)
datasets['default']=datasets['default'].map(category)
datasets['housing']=datasets['housing'].map(category)
datasets['loan']=datasets['loan'].map(category)
datasets['contact']=datasets['contact'].map(category)
datasets['poutcome']=datasets['poutcome'].map(category)
```

```python
datasets.info()
```

```python
X = datasets.drop(['job','marital','education','month','day_of_week','cons.conf.idx'],axis=1)
Y = datasets.iloc[:, 20].values

# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.20, random_state = 0)

# Fitting the classifier into the Training set

from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_Train, Y_Train)

# Predicting the test set results

Y_Pred = classifier.predict(X_Test)

# Making the Confusion Matrix 

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_Test, Y_Pred)
print(cm)
print("Accuracy of SVM:",metrics.accuracy_score(Y_Test, Y_Pred))
```
